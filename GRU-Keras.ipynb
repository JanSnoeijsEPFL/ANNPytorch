```python
import keras
import keras.layers as layers
from keras.models import Sequential
from keras.layers.wrappers import TimeDistributed
from keras import backend as K
import numpy as np

SeqL=10
NbSeq=4
np.random.seed(3)
X_train = np.random.randint(0, 2, [NbSeq, SeqL])
print(X_train.shape)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

X_test =  np.random.randint(0, 2, [NbSeq, SeqL])
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
                     
T_train = np.empty([NbSeq, SeqL])
T_train = np.reshape(T_train, (T_train.shape[0], T_train.shape[1], 1))
T_test = np.empty([NbSeq, SeqL])
T_test = np.reshape(T_test, (T_test.shape[0], T_test.shape[1], 1))
for k in range(0,SeqL):
    indexes = np.array([np.arange(0,k+1)])
    T_train[:,k] = X_train[:,0:k].sum(axis=1).astype(int)
    T_test[:, k] = X_test[:,0:k].sum(axis=1).astype(int)
    #T_train[:,k] = X_train[:,0:k].sum(axis=1).astype(int).squeeze()
    #T_test[:, k] = X_test[:,0:k].sum(axis=1).astype(int).squeeze()
#T_train=T_train[:,SeqL-1]
#T_test=T_test[:,SeqL-1]

#declare model
model = Sequential()
model.add(layers.GRU(units=1, input_shape=(SeqL,1), activation='tanh', recurrent_activation='hard_sigmoid',
              return_sequences=True))

model.add(layers.Dense(units=1, activation='relu'))
model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['accuracy'])

model.fit(X_train, T_train, batch_size=1, epochs=50, verbose=1)
print(model.predict(X_test).squeeze(), T_test.squeeze())
print(model.predict(X_train).squeeze(), T_train.squeeze())
score = model.evaluate(X_test, T_test, verbose=0)
```
